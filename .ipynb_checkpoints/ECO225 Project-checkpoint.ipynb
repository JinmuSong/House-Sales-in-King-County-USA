# Influencing Factors And Forecast Of Price Of Houses In King County, USA

```python
# Import Packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Introduction

The housing market is an important part of the economy. However, assessing the value of a property in an area is a relatively complex process, because the value of a property is usually a combination of various factors and conditions of the property. In recent years, the real estate market in King County, USA has experienced significant growth, which has attracted the attention of researchers and policy makers. Understanding the factors affecting home prices in King County and developing accurate forecasting models are critical to ensuring sustainable growth in the housing market. The purpose of this project is to explore the main factors affecting the housing price in King County, and develop the housing price forecast model. This project will use statistics and data analysis to study housing prices and related factors.

The primary data is obtained from kaggle.com (https://www.kaggle.com/datasets/harlfoxem/housesalesprediction/discussion/207885?resource=download), which provides data on more than 21,000 listed King County properties, along with their related property information (such as number of bedrooms and bathrooms, living area, number of floors, and house level).

After a lot of data processing and analysis and linear regression, we found that these relevant factors all affect the value assessment of real estate to a certain extent. This project selectively selected five factors with a high degree of correlation with real estate value (number of bedrooms and bathrooms, living area, number of floors, and house level) as independent variables and analyzed the degree of correlation between them and housing price.



## Variables

**Y Variable:** 
1. price: Price of the house in million dollars

**X Variable:** 
1. bedrooms: Number of bedrooms in the house
2. bathrooms: Number of bathrooms in the house
3. sqft_living: Size of the living area of the house in square feet
4. floors: Number of floors in the house
5. grade: Grade of the house determined by construction and design from 1 to 13

Independent variables above are chosen because they themselves are important factors in examining a property. The number of bedrooms and bathrooms determines the number of families a house can accommodate, the size of the living space and the number of floors determines the range of activities of the occupants in the house, and the grade of the house reflects the degree of good construction and design of the property.

Independent variables I chose are usually positively correlated with the overall quality of the property and the living experience, and the dependent variable I choose - price - is often a quantitative representation of the overall quality of a property, so these independent variables are most likely to be positively correlated with my dependent variable as well.

These influencing factors and their ability to influence will be analyzed and discussed in more detail in this report, after which the model will be used to forecast the value of the newly listed properties after extensive data analysis practice, where one can provide information about the property (e.g. number of bedrooms and bathrooms) to get a relatively accurate forecast of the offer.

## Data Loading And Cleaning

```python
# Read File
file = pd.read_csv(r"C:\Users\sjm28\Documents\ECO225\Project Repository\kc_house_data.csv")

# Change unit of house prices
file['price_MM'] = file['price'] / 1000000

# Extract factor columns
data = file[['price_MM','bedrooms','bathrooms','sqft_living','floors','grade',]]
data_x = file[['bedrooms','bathrooms','sqft_living','floors','grade',]]
```

## Plots, Histograms, Figures

### Box plot

The box plot basically provide the same information to the summary statistics.

```python
# Create figure and plot
fig, ax = plt.subplots(2, 2,figsize = (10,6))

#Adjust color
bgcolor = (250/255, 250/255, 250/255)
fig.set_facecolor(bgcolor)

#Adjust the space between plots
plt.subplots_adjust(wspace = 2, hspace = 0.4)

#plot
data_x.plot(kind = 'box', y = 'bedrooms', ax = ax[0][0], ylim = [0, 13], label = 'Number of Bedrooms')
data_x.plot(kind = 'box', y = 'bathrooms', ax = ax[0][1], label = 'Number of Bathrooms')
data_x.plot(kind = 'box', y = 'floors', ax = ax[1][0], label = 'Number of Floors')
data_x.plot(kind = 'box', y = 'grade', ax = ax[1][1], label = 'Grades')

#remove spines
for (i,col) in enumerate(data_x[['bedrooms', 'bathrooms', 'floors', 'grade']]):
    if i <= 1:
        x = 0
    else:
        x = 1
    if i % 2 == 0:
        y = 0
    else:
        y = 1
    ax[x][y].spines['right'].set_visible(False)
    ax[x][y].spines['top'].set_visible(False)
    ax[x][y].set_facecolor(bgcolor)

#set title
ax[0][0].set_title('Distribution of Number of Bedrooms of Houses In King County, USA')
ax[0][1].set_title('Distribution of Number of Bathrooms of Houses In King County, USA')
ax[1][0].set_title('Distribution of Number of Floors of Houses In King County, USA')
ax[1][1].set_title('Distribution of Grades of Houses In King County, USA')
```

### Histogram

The six histograms above show the distribution of different characteristics of a house, such as the number of bedrooms, living area, etc. On the x-axis of the image are the different characteristics of the house, and on the y-axis is the density.
Roughly speaking, there are four distributions that are positively skewed, namely the number of bathrooms, the size of the living area, the number of floors, and the price of the house. The remaining two, which can be considered approximately normal, are the number of bedrooms and the class of the house, respectively.

```python
# Create figure and plot
fig, ax = plt.subplots(6,1,figsize = (6,20))

#Adjust color
bgcolor = (250/255, 250/255, 250/255)
fig.set_facecolor(bgcolor)

#Adjust the space between plots
plt.subplots_adjust(hspace = 0.35)
data.hist(column = 'bedrooms', ax = ax[0], range = [0, 11], bins = 11, density = True)
data.hist(column = 'bathrooms', ax = ax[1], bins = 8, density = True)
data.hist(column = 'sqft_living', ax = ax[2], range = [0, 10000], bins = 100, density = True)
data.hist(column = 'floors', ax = ax[3], bins = 5, density = True)
data.hist(column = 'grade', ax = ax[4], bins = 12, density = True)
data.hist(column = 'price_MM', ax = ax[5], range = [0, 4], bins = 100, density = True)

#set y label and removed spines
for (i,col) in enumerate(data.columns):
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['top'].set_visible(False)
    ax[i].set_ylabel("Density".format(col))
    ax[i].set_facecolor(bgcolor)

#set x label and title
ax[0].set_xlabel('Number of Bedrooms')
ax[1].set_xlabel('Number of Bathrooms')
ax[2].set_xlabel('Living Area Size (ft^2)')
ax[3].set_xlabel('Number of Floors')
ax[4].set_xlabel('Grades')
ax[5].set_xlabel('Price (million$)')
ax[0].set_title('Distribution of Number of Bedrooms of Houses In King County, USA')
ax[1].set_title('Distribution of Number of Bathrooms of Houses In King County, USA')
ax[2].set_title('Distribution of Living Area Size(ft^2) of Houses In King County, USA')
ax[3].set_title('Distribution of Number of Floors of Houses In King County, USA')
ax[4].set_title('Distribution of Grades of Houses In King County, USA')
ax[4].set_title('Distribution of Price of Houses In King County, USA')
```

## Relation scatterplot
1. Number of Bedrooms: normal distribution, price highest at 5 - 6 bedrooms
2. Number of Bathrooms: positively correlated, more bathrooms, higher the price
3. Living area: positively correlated, larger living area, higher the price
4. Number of floors: price highest at houses with 2 floors
5. Grades: positively correlated, higher the grade of the house, higher the price

```python
seq0 = list(range(0, 12))
seq4 = list(range(1, 14))

# Create figure and plot
fig, ax = plt.subplots(5,1,figsize = (6,20))

#Adjust color
bgcolor = (250/255, 250/255, 250/255)
fig.set_facecolor(bgcolor)

#Adjust the space between plots
plt.subplots_adjust(hspace = 0.35)

#plot
for (i,col) in enumerate(data_x.columns):
    data.plot(kind = 'scatter', x = col, y = 'price_MM', ax = ax[i])
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['top'].set_visible(False)
    ax[i].set_ylabel('Price in million') # set y lable
    ax[i].set_facecolor(bgcolor)
    
#set x limit, x ticks, x label and title
ax[0].set_xlim((-.5, 12))
ax[0].set_xticks(seq0)
ax[2].set_xlim((-100, 10000))
ax[4].set_xlim((.5, 13.5))
ax[4].set_xticks(seq4)
ax[0].set_xlabel('Number of Bedrooms')
ax[1].set_xlabel('Number of Bathrooms')
ax[2].set_xlabel('Living Area Size (ft^2)')
ax[3].set_xlabel('Number of Floors')
ax[4].set_xlabel('Grades')
ax[0].set_title('Scatterplot of Price(million$) vs Number of Bedrooms of Houses In King County, USA')
ax[1].set_title('Scatterplot of Price(million$) vs Number of Bathrooms of Houses In King County, USA')
ax[2].set_title('Scatterplot of Price(million$) vs Living Area Size(ft^2) of Houses In King County, USA')
ax[3].set_title('Scatterplot of Price(million$) vs Number of Floors of Houses In King County, USA')
ax[4].set_title('Scatterplot of Price(million$) vs Grades of Houses In King County, USA')
```

## Relation between average living area size and grade

The bar chart below shows the average living area for different grade of housing. The Y-axis is the average living area, measured in square feet. The X-axis is the grade of the house, on a scale from 1 to 13. We can observe a clear upward trend, as the house grade gets higher, the living area also increased.

```python
file['avg_sqft'] = file.groupby('grade')['sqft_living'].transform('mean')
avg_sqft_grade = file[['avg_sqft','grade']].drop_duplicates(subset = 'grade').sort_values(by = 'grade')

# Create figure and plot
fig, ax = plt.subplots()

#Adjust color
bgcolor = (250/255, 250/255, 250/255)
fig.set_facecolor(bgcolor)

#plot
avg_sqft_grade.plot.bar(x = 'grade', y = 'avg_sqft', ax = ax, legend = False)

# remove spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#set title, y label and x label
ax.set_title("Average Living Area Size of Houses In Different Grades")
ax.set_ylabel('Average Living Area Size(ft^2)')
ax.set_xlabel('Grade')
ax.set_facecolor(bgcolor)
```
